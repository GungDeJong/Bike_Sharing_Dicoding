## Visualization & Explanatory Analysis

### Bagaimana pengaruh musim terhadap jumlah total rental sepeda pada tahun 2011 dan 2012

```python
#Pembuatan DF berdasarkan Tahun 2011 dan 2012
bsH_2011 = bsH[bsH['yr'] == 2011]
bsH_2012 = bsH[bsH['yr'] == 2012]
```

```python
bsH_2012['yr'].unique()
```

```python
# Pembuatan DF baru dengan feature season dan cnt berdasarkan tahun 2011
new_bsH_2011 = bsH_2011[['season', 'cnt']].copy()
new_bsH_2011.head()
```

```python
# Pembuatan DF baru dengan feature season dan cnt berdasarkan tahun 2011
new_bsH_2012 = bsH_2012[['season', 'cnt']].copy()
new_bsH_2012.head()
```

```python
fig, axes = plt.subplots(nrows = 2, ncols = 1, figsize = (10,6))

#Berdasrkan Tahun 2011
sns.barplot(
    x = 'season',
    y = 'cnt',
    data = new_bsH_2011,
    ax = axes[0]
)
axes[0].set_title("Total Rental Setiap Season Berdasrkan Tahun 2011")
axes[0].set_xlabel("Season")
axes[0].set_ylabel("Total Rental")

#Berdasarkan 2012
sns.barplot(
    x = 'season',
    y = 'cnt',
    data = new_bsH_2012,
    ax = axes[1]
)
axes[1].set_title("Total Rental Setiap Season Berdasrkan Tahun 2012")
axes[1].set_xlabel("Season")
axes[1].set_ylabel("Total Rental")

plt.tight_layout()
plt.show()
```

### Bagaiamana pola total rental setiap bulannya pada tahun 2011 dan 2012

```python
bsH['mnth'] = pd.Categorical(bsH['mnth'], categories = ['Jan', 'Feb', 'Mar', 'Apr', 'Mei', 'Jun', 'Jul', 'Ags', 'Sep', 'Okt', 'Nov', 'Des'], ordered = True)

total_mnth = bsH.groupby(by = ['mnth', 'yr']).agg({
    'cnt' : 'sum'
}).reset_index()

sns.lineplot(
    x = 'mnth',
    y = 'cnt',
    hue = 'yr',
    marker = "o",
    data = total_mnth
)

plt.title("Pola Total Rental Sepeda Setiap Bulan Pada Tahun 2011 dan 2012")
plt.xlabel(None)
plt.ylabel(None)
plt.legend(title = "Year", loc ="lower right")
plt.tight_layout()
plt.show()
```
